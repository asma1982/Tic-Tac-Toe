```python
def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_win(board, symbol):
    for i in range(3):
        if all(board[i][j] == symbol for j in range(3)):
            return True
        if all(board[j][i] == symbol for j in range(3)):
            return True
    return all(board[i][i] == symbol for i in range(3)) or all(board[i][2 - i] == symbol for i in range(3))

def check_tie(board):
    return all(all(square != "" for square in row) for row in board)

def play_game():
    board = [["" for _ in range(3)] for _ in range(3)]
    players = ["X", "O"]
    current_player = 0

    while True:
        print_board(board)
        print(f"دور اللاعب {players[current_player]}.")

        row = int(input("ادخل الصف (0-2): "))
        col = int(input("ادخل العمود (0-2): "))

        if board[row][col] == "":
            board[row][col] = players[current_player]

            if check_win(board, players[current_player]):
                print_board(board)
                print(f"اللاعب {players[current_player]} يفوز!")
                break
            elif check_tie(board):
                print_board(board)
                print("تعادل!")
                break

            current_player = 1 - current_player
        else:
            print("هذا الموقع مشغول. حاول مرة أخرى.")

play_game()
```
